```python
import pandas as pd
import numpy as np
```

```python
loan = pd.read_csv('/content/drive/MyDrive/loan_data.csv')
```

```python
np.subtract(loan['customer_id'].unique().shape, loan.shape[0])
```

```python
loan['customer_id'].duplicated().sum()
```

```python
dict(loan.dtypes)
```

```python
loan.set_index('customer_id' , inplace=True)
```

```python
loan.head()
```

```python
loan.sort_values(by="customer's_nth_loan", inplace=True, ascending=False)
```

```python
loan['sign_date']=pd.to_datetime(loan['loan_signdate'])
```

```python
loan.drop(loan.index[7631], axis=0, inplace=True)
```

```python
loan['loan_contractapproveddate'].replace('0000-00-00 00:00:00', '', inplace=True)
```

```python
loan['approved_date'] = pd.to_datetime(loan['loan_contractapproveddate'])
```

```python
loan['sign_month'], loan['approved_month']=(loan['sign_date'].dt.month_name(), loan['approved_date'].dt.month_name())
```

```python
loan['signed_year'], loan['approved_year']=loan['sign_date'].dt.year, loan['approved_date'].dt.year
```

```python
loan.dropna(axis=0, inplace=True)
```

```python
loan[['approved_year','signed_year']]=loan[['approved_year','signed_year']].astype('int')
```
